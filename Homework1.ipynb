#### Question 4 (d)

SE($\hat{\beta}$) = $\frac{\sqrt{\sum(y_i^2-2x_iy_i\hat{\beta}+\hat{\beta}^2x_i^2)}}{\sqrt{(n-1)\sum(x_i^2)}}$\
\
 = $\frac{\sqrt{\sum(y_i^2-2\frac{(x_iy_i)^2}{x_i^2}+\frac{(x_iy_i)^2}{x_i^2})}}{\sqrt{(n-1)\sum(x_i^2)}}$ \
 \
 = $\frac{\sqrt{\sum(y_i^2-\frac{(x_iy_i)^2}{x_i^2})}}{\sqrt{(n-1)\sum(x_i^2)}}$

$\hat{\beta} = \frac{\sum_{i=1}^{n} x_i y_i}{\sum_{i=1}^{n} x_i^2}$ \
\
$\hat{\beta}$/ SE($\hat{\beta}$) = $\frac{{\sum_{i=1}^{n} x_i^2}}{\sum_{i=1}^{n} x_i y_i}$ $\frac{\sqrt{\sum(y_i^2-\frac{(x_iy_i)^2}{x_i^2})}}{\sqrt{(n-1)\sum(x_i^2)}}$\
\
= $\frac{\sqrt{(n-1)}\sum(x_iy_i)}{\sqrt{(\sum{x_i^2})\sum(y_i^2-\frac{(x_iy_i)^2}{x_i^2})}}$\
\
= $\frac{\sqrt{(n-1)}\sum(x_iy_i)}{\sqrt{(\sum{x_i^2})(\sum{y_i^2})-(\sum{x_iy_i})^2}}$

```python
# Question 4 (d)
rng = np.random.default_rng(1)
x = rng.normal(size=100)
y = 2 * x + rng.normal(size=100)

n = len(x)
xy = np.sum(x * y)
x2 = np.sum(x ** 2)
y2 = np.sum(y ** 2)

(np.sqrt(n-1) * xy) / np.sqrt((x2 * y2) - (xy**2))
```
